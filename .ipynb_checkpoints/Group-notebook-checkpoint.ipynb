Metric 1: The first metric we will examine is the Police Firearm Seizures in Pittsburgh neighborhoods. The data describes what firearm was seized by the Pittsburgh Bureau of Police and describes the type of incident that occurred. The neighborhood that had the most firearm seizures was Homewood South with 345, followed by South Side Flats with 208. From this, we can tell that these areas had the biggest issue when it came to firearms posing a threat to individuals within the community. Firearm seizures can also refer to the firearms that were taken because of an illegal purchase. This means that these areas also have an issue when it comes to weapons being purchased illegally or without a proper warrant. 

```python
import pandas as pd
import geopandas
%matplotlib inline

seizures = pd.read_csv("police_firearms.csv")

num_seizures = seizures.groupby("neighborhood").sum()['total_count']
num_seizures.sort_values(ascending=False)

neighborhoods = geopandas.read_file("Neighborhoods/Neighborhoods_.shp")
neighborhoods.plot()

seizures_map = neighborhoods.merge(num_seizures, how='left', left_on='hood', right_on='neighborhood')
seizures_map[['hood','total_count','geometry']].head()
seizures_map.plot(column='total_count')
seizures_map.plot(column='total_count',
               cmap='PuBu',            
               edgecolor="black",   
               legend=True,        
               legend_kwds={'label': "Total Firearm Seizures"},
               figsize=(15, 10),   
               missing_kwds={"color": "white"}
               )

```

Metric 2: The second metric evaluates the frequency of fire incidents in Pittsburgh neighborhoods. The database describes what kind of fire incident has occurred, the area, and the alarm time. The Neighborhood that was least common in the dataset and had the least fire incidents can be considered the safest and the most protected from a fire incident. According to the graph. Mount Oliver borough and Regent Square are the safest areas of the city. On the other hand, Carrick and Central Business District experienced the biggest amount of fire incidents. This metric can help us to evaluate trends, assessments, and risks that Pittsburgh neighborhoods experience. 


```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
file_path = 'Fire_project.csv'

# Creating hte database
df = pd.read_csv(file_path)
df.head()
```

```python
# Receiving the values of the neighborhod columns
neighborhood_counts = df['neighborhood'].value_counts()

# Top 40 least and most common values
top_40 = neighborhood_counts.head(40)
least_40 = neighborhood_counts.tail(40)

# Graphing 40 most common values
plt.figure(figsize=(12, 6))
top_40.plot(kind='bar', color='skyblue')
plt.title('Top 40 Most Common Neighborhoods')
plt.xlabel('Neighborhood')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('top_worst40.pdf')
plt.show()



# Graphing top 40 least common values
least_40 = least_40.iloc[::-1]
plt.figure(figsize=(12, 6))
least_40.plot(kind='bar', color='lightcoral')
plt.title('Top 40 Least Common Neighborhoods')
plt.xlabel('Neighborhood')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('top_best40.pdf')
plt.show()
```

```python
print("Top 40 Most Common Neighborhoods:")
print("{:<30} {:<10}".format('Neighborhood', 'Count'))
for neighborhood, count in zip(top_40.index, top_40.values):
    print("{:<30} {:<10}".format(neighborhood, count))
#table for top 40 least common neighborhoods
print("\nTop 40 Least Common Neighborhoods:")
print("{:<30} {:<10}".format('Neighborhood', 'Count'))
for neighborhood, count in zip(least_40.index, least_40.values):
    print("{:<30} {:<10}".format(neighborhood, count))
```

Based on the Fire incidents metric, neighborhood that appears in the database least amount of times is considered the safest and most protected from fire incidents. Neighborhoods that appear more freqeuntly are more lickely to experience a fire incident. According to the graph. Mount Oliver borough and Regent square are the safest neighborhoods.
