The first metric we will examine is the Police Firearm Seizures in Pittsburgh neighborhoods. The data describes what firearm was seized by the Pittsburgh Bureau of Police and describes the type of incident that occurred. The neighborhood that had the most firearm seizures was Homewood South with 345, followed by South Side Flats with 208. From this, we can tell that these areas had the biggest issue when it came to firearms posing a threat to individuals within the community. Firearm seizures can also refer to the firearms that were taken because of an illegal purchase. This means that these areas also have an issue when it comes to weapons being purchased illegally or without a proper warrant. 


The map below shows the trends in where the most firearm seizures are in the city.

```python
import pandas as pd
import geopandas
%matplotlib inline

seizures = pd.read_csv("police_firearms.csv")

num_seizures = seizures.groupby("neighborhood").sum()['total_count']
num_seizures.sort_values(ascending=False)

neighborhoods = geopandas.read_file("Neighborhoods/Neighborhoods_.shp")
neighborhoods.plot()

seizures_map = neighborhoods.merge(num_seizures, how='left', left_on='hood', right_on='neighborhood')
seizures_map[['hood','total_count','geometry']].head()
seizures_map.plot(column='total_count')
seizures_map.plot(column='total_count',
               cmap='PuBu',            
               edgecolor="black",   
               legend=True,        
               legend_kwds={'label': "Total Firearm Seizures"},
               figsize=(15, 10),   
               missing_kwds={"color": "white"}
               )

```
