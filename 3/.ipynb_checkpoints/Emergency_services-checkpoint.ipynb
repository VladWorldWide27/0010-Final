```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
file = '3rdmetric.csv'

# Creating the database
df = pd.read_csv(file)
df.head()

#Making the Data from district and area of medical service
df = df[["area", "district"]]
df.plot.bar()
plt.title('Area of Medical Serrvices in Pittsburgh Districts')
plt.xlabel('Districts')
plt.ylabel('Area')
```

Due to recent changes the districts after 9 have all been placed inside district 9 meaning district 9, 5, 4, and 7 have the most medical service area. This is a metric that can consider how safe a neighborhood is based on the size of the medical service area and the districts with the most are 9, 5, 4, and 7 so those neihborhoods in the distritcs are considered safe. They include: District 5 (Glen Hazel, Greenfield, Hays, Hazelwood, Lincoln Place, New Homestead, Regent Square, Squirrel Hill South,  and Swisshelm Park), District 4 (Beechview, Bon Air, Brookline, Carrick, Mt. Washington, and Overbrook), District 7 (Bloomfield. Friendship, Highland Park, Lawrencville, Morning Side, Polish Hill, Stanton Heights, and Strip District), and finally District 9 (East Liberty, Homewood, East Hills, Larimer, Point Breeze North, Lincoln-Lemington-Belmar, and Garfield). These are the neighborhoods with the most safety with District 5 having the most out of them all. The least safe District is 8. This district consists of the neighborhoods Oakland, Shadyside, and Squirrel Hill.
